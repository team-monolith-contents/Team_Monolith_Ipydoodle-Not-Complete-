## **거울을 직접 구현해보기**

#### Step 2. 방 만들기

먼저 영희의 방을 임의로 만드는 함수 `make_room` 함수를 만들어봐요.
* 식탁은 상자로 만들고 화분은 원으로 표시합니다. 
* `make_room` 함수를 호출할때마다 식탁과 화분의 크기와 위치는 랜덤하게 바뀌게 해주세요.
* 방을 반으로 나누었을 때 화분은 x축을 기준으로 아래 쪽에, 식탁은 위 쪽에 있도록 해주세요. 화분과 식탁은 각각의 영역을 벗어나서도, 겹쳐서도 안돼요.
* 도전!) 색깔도 랜덤하게 설정해봅시다. 색깔은 길이 3짜리 배열이고, 각 배열의 원소는 0 이상 255 이하로 표현해줍니다.

화분과 식탁과의 위치와 크기, 색깔을 랜덤하게 설정해야하는데, 어떻게 할까요? 파이썬의 `random` 라이브러리를 사용해줄께요.   
아래의 예시 코드를 사용하면 범위 안에서 임의의 정수를 뽑아낼 수 있어요. 첫번째 인자는 최소값, 두번째 인자는 최대값을 의미해요.

```python
# 예제 1. random 함수 활용 예시

from random import randint # random 라이브러리에서 randint함수를 import해옵니다. 
for i in range(5):
    random_number = randint(1, 10) # 1 이상 10 이하의 정수를 뽑아줘!
    print(random_number)
```


위에서 배운 `randint` 함수를 활용하여 가구를 배치해봅시다. (최대한 아래의 가이드라인을 보지 않고 만들어보세요.)
1. World() 함수로 공간을 만든 후, Line() 함수로 방의 경계를 만들 수 있습니다. 
2. 방의 경계가 만들어지면, 그 경계 내부에 desk와 flowerpot의 위치를 랜덤하게 지정해줍니다. 
3. desk가 방의 경계를 넘지 않게 하려면, desk의 좌표가 높이와 넓이의 최대값의 절반만큼 방의 경계에서 떨어져야 합니다.
4. 마찬가지로 flowerpot이 방의 경계를 넘지 않게 하려면, flowerpot의 좌표가 반지름만큼 방의 경계에서 떨어져야 합니다. 
5. 방의 경계를 넘지 않도록 가구들의 위치 좌표값을 랜덤하게 설정하는데 유의하세요. 

```python
# 예제 2. Ipydoodle에 randint 사용하기
from random import randint

def make_room(): # make_room() 함수를 선언해줍니다. 
    room = World(width = 700, height = 700)
    Line(x1 = -350, y1 = 350, x2 = 0, y2 = 350)
    Line(x1 = -350, y1 = -350, x2 = -350, y2 = 350)
    Line(x1 = -350, y1 = -350, x2 = 0, y2 = -350)
    # 위의 코드는 수정하지 말아주세요! 
    desk = Box(x = randint(-300,-50), y = randint(50,300), width = randint(50,100), height = randint(50,100), color = [randint(0,255),randint(0,255),randint(0,255)])
    flowerpot = Circle(x = randint(-250,-100), y = randint(-250,-100), radius = randint(50,100), color = [randint(0,255),randint(0,255),randint(0,255)])

    # desk = ...
    # flowerpot = ...

    return desk, flowerpot

desk, flowerpot = make_room()
```

#### Step 3. 거울 그리기

이제 주어진 방에 거울을 설치해봅시다. 영희는 거울이 큰걸 좋아하기 때문에, 방의 가장 긴 면에 설치하려고 해요

```python
#예제 3 Line 함수를 활용하여 거울을 그려봅시다!
desk, flowerpot = make_room()
Line(x1 = 0, y1 = -350, x2 = 0, y2 = 350, color = 'blue')
# Line = ...
```

#### Step 4.거울에 물체 비추기

앞서 말했듯, 영희의 방에는 두가지 가구가 있어요. 네모 모양의 `desk`와 원 모양의 `flowerpot`이 있습니다.  
거울에 비친 두 가구의 모습을 만들기 위해서는 다음의 문제를 풀어야 해요.

**1. 방에 있는 두 가구의 위치 가져오기**  
**2. 거울에 비친 두 가구의 위치 계산하기**  
**3. 계산한 위치에 똑같은 모습의 두 가구 만들기**

#### Step 4-1. 방에 있는 두 가구의 위치 가져오기  
`make_room`은 `desk`, `flowerpot`을 반환합니다. `desk`와 `flowerpot`로부터 위치를 뽑아볼까요?  
각각의 변수로부터 `pos`함수를 호출하면 위치를 계산할 수 있습니다.   
뽑아낸 위치 정보를 배열 형태로 각각의 다른 변수에 저장해주세요. 
* `pos`를 호출해 받은 변수는 `tuple`이라는 형태의 다른 변수입니다. 이를 배열로 변환하려면 어떻게 해야할까요? 

```python
# 예제 4-1
# print(desk.pos)
# print(flowerpot.pos)
desk_pos = list(desk.pos) # list() 함수를 사용하여 tuple 변수를 배열 변수로 변환합니다. 
flowerpot_pos = list(flowerpot.pos)
print(desk_pos, flowerpot_pos)
```

#### Step 4-2 거울에 비친 두 가구의 위치 계산하기  
거울은 방의 가장 긴 면에 만들겁니다. 좌표상의 y축과 같죠.  
그렇다면 거울에 비친 두 가구의 위치는 어떻게 될까요? (힌트: 위에 있는 거울속의 원리를 참조하세요)
정답! 거울에 비친 두 가구는 원래의 가구와 y축 선대칭을 이루게 되겠죠.
y축 선대칭을 만들기 위해서는 앞서 구한 pos변수를 처리해야 합니다.   
pos는 배열이었죠? 배열을 처리하는 방법을 먼저 배워볼게요.  

#### <배열의 Index 활용하기>
desk_pos와 flowerpot_pos는 길이 2짜리 배열입니다! (첫번째 원소는 x좌표이고, 두번째 원소는 y좌표에요.)  
배열에 있는 원소를 편집하려면 어떻게 해야할까요?   
배열의 원소는 인덱스 번호를 활용하여 편집합니다. 다음 예시 코드를 보고 인덱스 번호를 알아봅시다.   
(인덱스 코드는 0부터 시작합니다)

```python
#예제 5 index 번호 활용하기
a = [1, 2, 3] # 배열에 저장된 원소 각각에 0부터 순서대로 인덱스 번호가 붙습니다. (예를 들어, 인덱스 번호 1에 저장된 값은 2입니다.)
print('a[0]:', a[0]) # 각 인덱스에 저장된 값에 호출하려면 대괄호([])를 사용해야 합니다. a[0]과 같이 "배열이름[인덱스]"의 형태로 호출합니다. 
a[2] = 4 # 배열 array의 2번 칸에 -1을 저장. 이처럼 인덱스로 호출한 값을 변경할 수도 있습니다. 
print('a[2]:', a[2])
a[0] = a[1] # 배열 array의 0번 칸에 1번칸의 값을 저장. 좌변의 위치에 우변의 값을 저장합니다.(변수에 자료를 저장하는 것과 동일한 방법이죠?)  
print('a[0]:', a[0])
```

위에서 배운 배열과 인덱스의 개념을 사용해서 `desk_pos`와 `flowerpot_pos`의 값을 y축 선대칭이 되도록 편집해보세요

```python
# 예제 6 y축을 기준으로 선대칭을 시키기.(y좌표는 유지되고, x좌표는 부호가 바뀔거에요)
# 인덱스 복습! 첫번째 원소인 x좌표값을 변경하려면 0번째 인덱스를 호출해야겠죠?
desk_pos[0] = -desk_pos[0] # desk_pos[0]은 desk의 x좌표입니다. 이에 "-"를 붙여 음수로 만들면 선대칭이 됩니다.(desk_pos[0] *= -1)
flowerpot_pos[0] = -flowerpot_pos[0] # desk와 동알한 방식으로 x좌표를 호출하고 음수로 변경해줍니다.(flowerpot_pos[0] *= -1)
print(desk_pos, flowerpot_pos)
```

#### Step 4-3 계산한 위치에 똑같은 모습의 두 가구 만들기  
계산한 위치에 각각 두 가구를 만들어볼까요? 참고로 영희가 가지고 있는 가구들은 거꾸로 봐도 똑바로 봐도 똑같답니다. 
> 주의: 가구를 복사하면 처음 `make_room`을 호출했던 위치에 가구가 생길거에요.  
 

       
가구들의 색깔과 크기는 다음을 통해 가져올 수 있어요.  
* `desk.width`: 식탁의 가로 길이
* `desk.height`: 식탁의 세로 길이
* `desk.color`: 식탁의 색깔
* `flowerpot.radius`: 화분의 반지름
* `flowerpot.color`: 화분의 색깔

```python
#예제 7 거울로 비친 값을 desk_copy, flowerpot_copy에 저장하기
desk_copy = Box(x = desk_pos[0], y = desk_pos[1], color = desk.color, width = desk.width, height = desk.height)
flowerpot_copy = Circle(x = flowerpot_pos[0], y = flowerpot_pos[1], color = flowerpot.color, radius = flowerpot.radius)
```

#### Step 5 최종값 출력!
이 때까지 한 것들을 한번 모아서 한번에 실행해 볼까요?

```python
desk, flowerpot = make_room() # world, 방, 가구 만들기
Line(x1 = 0, y1 = -350, x2 = 0, y2 = 350, color = 'blue') #거울 만들기

desk_pos = list(desk.pos) # tuple을 list로 변환
flowerpot_pos = list(flowerpot.pos) # tuple을 list로 변환

desk_pos[0] = -desk_pos[0] # y축 선대칭
flowerpot_pos[0] = -flowerpot_pos[0] # y축 선대칭

desk_copy = Box(x = desk_pos[0], y = desk_pos[1], color = desk.color, width = desk.width, height = desk.height)
flowerpot_copy = Circle(x = flowerpot_pos[0], y = flowerpot_pos[1], color = flowerpot.color, radius = flowerpot.radius)
```
