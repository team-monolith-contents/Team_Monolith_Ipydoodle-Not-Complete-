## 0. Intro

자판기는 알맞은 금액을 넣고 원하는 물건을 선택하면 그 물건이 나오는 일종의 무인판매기입니다.  
영희는 이 자판기에서 가능성을 보았고, 이제는 이 자판기를 직접 만들어 사업을 하려고 합니다.  
여러분은 영희에게 이 자판기에 들어갈 프로그램을 만들어달라고 의뢰를 받은 상황입니다.  
여러분이 영희에게 만들어주기로 한 자판기는 다음처럼 행동합니다

* 처음 자판기를 고르면 음식을 골라야 한다.
* 음식 이름을 잘못 입력하면 오류가 난다.
* 음식 이름을 알맞게 입력하면 그 음식을 출력한다.

영희가 의뢰한 자판기 프로그램을 개발해주세요.

### 2.2. 음식 이름 비교하기

다음을 만족하는 프로그램을 작성해주세요.

### 2.3. 음식 이름 바탕으로 음식 출력해주기

앞서 예제 1.3에서 우리가 만든 음식은 두가지였습니다. 음식을 입력받아 알맞은 음식을 출력할 수 있도록 해주세요.

```python
if ("조건1"):

    # 조건 2일때 실행할 코드

    # 조건 1, 2도 아닐때 실행할 코드
```

다음을 만족하는 프로그램을 작성해주세요.
1. 음식 이름을 입력으로 받습니다.
